```python
import matplotlib.pyplot as plt 
import numpy as np 
from skimage import io 
from sklearn.cluster import KMeans 
from sklearn.utils import shuffle
n_Quan = 2

main_img = io.imread("man.bmp")
img = np.array(main_img,  dtype = np.float64 )/255 # make easier computation
h,w = img.shape #grayscale image, we ignore "channle"
img_array = np.reshape(img, (h*w,1))
```

```python
img.shape
```

```python
#original image
plt.figure(1)
plt.clf()
plt.axis("off")
plt.title("Original Image")
plt.imshow(img,cmap = "gray")
####################################
#Histogram Original Image
plt.figure(2)
plt.clf()
plt.title("Histogram Original Image")
hist = np.histogram(main_img, bins=np.arange(0,256))
plt.plot(hist[1][:-1],hist[0], lw = 2)
```

```python
#train KMeans
img_array_sample = shuffle(img_array, random_state=0)[:1000]
kmeans = KMeans(n_clusters = n_Quan).fit(img_array_sample)
labels = kmeans.predict(img_array)
```

```python
#Creat the quantized Image
img_out = np.zeros((h,w))
label_idx = 0 
for i in range(h):
    for j in range(w):
        img_out[i][j] = kmeans.cluster_centers_[labels[label_idx]]
        label_idx += 1
```

```python
# Quantized Image
plt.figure(3)
plt.clf()
plt.axis("off")
plt.title("Quantized Image_2")
plt.imshow(img_out, cmap='gray')
#####################################
#Convert Quantized image for plotting 
img_out = np.array(img_out*255, dtype = np.uint8)

#Histogram Quantized Image
plt.figure(4)
plt.clf()
plt.title("Histogram Quantized Image_2")
hist = np.histogram(img_out, bins=np.arange(0,256))
plt.plot(hist[1][:-1],hist[0], lw = 2)
```
